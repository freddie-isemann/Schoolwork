# Assignment 3 - Write a `main` Method that Runs Elections

Add a `main` procedure to your class `Candidate_/username/` that runs elections using instances of the other Candidate classes. 

### About `main`

When a Java class has a public procedure (= static method) named 'main(String[])', 
the class can be run from the command line using `java` and the name of the class.
```
> java ClassName
[I/O produced by the 'main' method in ClassName]
>
```
The class A3 has no instance members, just a 'main' and some helper procedures, and no A3 objects are ever created.
It is possible to create A3 objects, 
but the type A3 acts as an empty class, equivalent to `class A3 { }`, 
because all its members are static.

Your class `Candidate_/username/` will have both a 'main' allowing the class to be 'run' 
and instance members (instance variables and instance methods) 
that allow objects of your class to have state and behaviour.

### Spec

Your main should offer the user the ability to add specific candidates to an election and the ability to run the same election many times and report the most common winner(s).
```
...
b

= Running Repeated Elections =
Candidates are
None
Would you like to a) add a specific candidate b) add a candidate at random c) run the election?
> a
Please enter a username.
> eey577
Candidates are
1. Ip
Would you like to a) add a specific candidate b) add a candidate at random c) run the election?
> c
Who should count the votes?
> eey577
How many times shall we run the election?
> 1000
Most common winner is Ip.
There were no other winners.

Would you like to 
a) exit 
b) run same election many times
c) check who counts honestly 
d) run all possible two-candidate elections
e) run a higher-order election
f) ...
```

Your elections should allow every `Candidate_/username/` object one vote and should offer the user a choice of who (which Candidate object) counts the votes.
