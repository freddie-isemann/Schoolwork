# Assignment 2

Write a class called `Candidate_/un/` in a file called `Candidate_/un/.java`, where `/un/` is your username.

Your class must be a sub-class of Candidate and so (see below) must implement four methods: 
`getName`, `vote`, `selectWinner` and `getSlogan`.

The abstract class Candidate and the interfaces Person, Citizen and Electable are given and cannot be changed.

```java
interface Person {
    String getName();
}
```

```java
interface Citizen extends Person {
    Candidate vote(Candidate[] candidates);
    Candidate selectWinner(Candidate[] votes);
} 
```

(An 'Unresolved dependencies' warning is OK because Candidate will be defined below. Just run the cell again after running Candidate to remove the warning.)

```java
interface Electable {
    String getSlogan();
}
```

```java
abstract class Candidate implements Citizen, Electable { }
```

## Spec

The string returned by your class's getName() method must not be longer than 15 characters.

The string returned by your getSlogan() must not be longer than 40 characters.

When passed an empty array, 
your vote(Candidate[]) and selectWinner(Candidate[]) methods must not throw an exception.


When passed a non-empty array, 
your vote(Candidate[]) and selectWinner(Candidate[]) must return an object from that array, so no write-in votes or surprise winners. (This will be tested.)

All your methods must run without any input/output activity: they must not print anything or consult the user.

Your methods must not take a long time or fail to return.

### Context 

The string returned by your getName() method will be used when an instance of your class is announced as the winner of an election. There is no need to use your real name, but it should be SFW.

The strings returned by your getName() and getSlogan() methods allow instances of your class to be identified and, possibly, attract votes. Your friends can write their vote methods to vote for instances of your class by recognizing your name or you can all vote for each other by recognizing a shared slogan.

Objects belonging to your class may be called upon to vote in an election or to count the votes in an election. 

Your vote method must somehow decide which object to return from the array of Candidate objects it is given. It should do this by calling the methods of those objects and considering the results. For example, it might search for an object in the array with a particular slogan and return that object. 

Your selectWinner(Candidate[]) method would normally return 
the most frequently occurring Candidate object in the array it is passed. 


### Examples

Here are some tame examples.

```java
class Candidate_ac1234 extends Candidate {
    
    public String getName() {
        return "Robin";
    }
    
    public String getSlogan() {
        return "More trees!";
    }
    
    public Candidate vote(Candidate[] candidates) {
        
        // If array is empty, return instance of friend's class.
        if (candidates.length == 0) 
            return new Candidate_ac1235();
        
        // Search for a like minded candidate.
        for (Candidate c : candidates)
            if (c.getSlogan().equals("More trees!"))
                return c;
        
        // Otherwise, search for a friend.
        for (Candidate c : candidates)
            if (c.getName().equals("Kim"))
                return c;
        
        // Otherwise, default to last candidate on list.
        return candidates[candidates.length-1];
    }
    
    public Candidate selectWinner(Candidate[] votes) {
        
        // If array is empty, return instance of friend's class.
        if (votes.length == 0) 
            return new Candidate_ac1235();
        
        // Default to first vote, but this will be over-written.
        Candidate currentWinner = votes[0];
        
        // Count the votes for each object in the array,
        // selecting one with the most.
        int highestCount = 0;
        for (Candidate c : votes) {
            
            int count = 0;
            for (Candidate x : votes)
                if (x == c)
                    count++;
            if (count > highestCount) {
                highestCount = count;
                currentWinner = c;
            }
        }
        
        return currentWinner;
    }
    
}
```

## A Main Program

Below is a simple example of a class with a main that runs an election.

You can use this to test your implementation, but - and hopefully this is obvious by now - it should not form part of your contribution. Each student contributes one file (Candidate_/un/.java) containing just the one class (Candidate_/un/).

The A2 repo will have a similar (but more complex) main class in a file A2.java, 
which calls all four methods in every student's contribution, 
runs a few tests on their behaviour, 
and allows the user to call various forms of election.

To develop your contribution, you are encouraged to get together with some friends, 
rewrite the above examples, modify the A2 below to match the names of your classes, 
and possibly modify it further to hold different elections.

Make sure you understand how this code works and ask/answer questions about it on the Module Forum.

```java
class A2 {
    public static void main(String[] args) {
        
        // Make up list of candidates to pass to vote methods.
        Candidate[] listOfCandidates = new Candidate[3];
        listOfCandidates[0] = new Candidate_ac1234();
        listOfCandidates[1] = new Candidate_ac1235();
        listOfCandidates[2] = new Candidate_ac1236();
        
        // Compile list of citzens.
        Citizen[] voters = new Citizen[3];
        voters[0] = new Candidate_ac1234();
        voters[1] = new Candidate_ac1235();
        voters[2] = new Candidate_ac1236();
        
        // Ask all voters to vote.
        Candidate[] ballotBox = new Candidate[voters.length];
        for (int i=0;i<voters.length;i++) {
            Candidate v = voters[i].vote(listOfCandidates);
            System.out.println(voters[i].getName() +" voted for " + v.getName() +
                              ", whose slogan is '" + v .getSlogan() + "'.");
            ballotBox[i] = v;
        }
        
        // Ask all voters to count the votes.
        for (int i=0;i<voters.length;i++) {
            Person winner = voters[i].selectWinner(ballotBox);
            System.out.println(voters[i].getName() +" says " + winner.getName() +
                              " wins the election.");
        }
        
    } 
}
```
